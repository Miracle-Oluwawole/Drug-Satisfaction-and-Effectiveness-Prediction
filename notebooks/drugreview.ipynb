# Data Preparation for Modelling

```python
#Text Preprocessing for 'Reviews' column
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
# Remove negatives from the stopword list so they stay in your text
negatives = {"no", "not", "nor", "never", "n't", "don", "aren", "couldn", "didn", "doesn", "hadn", "hasn", "haven", "isn", "wasn", "weren", "won", "wouldn"}
stop_words = stop_words - negatives
lemmatizer = WordNetLemmatizer()

#define function to clean text
def clean_text(text):
    if not isinstance(text, str):          ## Check if text is a string to handle NaNs
        return ""
    text = re.sub(r'[^a-zA-Z0-9]', ' ', text)   #keep letters and drug numbers
    text = text.lower()
    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words and len(word) > 1] #keep words with more than one letter
    return ' '.join(words)
```

```python
# Convert categorical columns
drugreview_df['Drug'] = drugreview_df['Drug'].astype('category')
drugreview_df['Condition'] = drugreview_df['Condition'].astype('category')
```

```python
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
```

```python
#initialise Vader Sentiment Analysis
sia = SentimentIntensityAnalyzer()
```

```python
# Create sentiment extraction functions

# TextBlob Sentiment
def get_textblob_polarity(text):
    return TextBlob(text).sentiment.polarity

def get_textblob_subjectivity(text):
    return TextBlob(text).sentiment.subjectivity


# VADER Sentiment
def get_vader_sentiment(text):
    score = sia.polarity_scores(text)
    return score['compound']     

def get_vader_negative(text):
    return sia.polarity_scores(text)['neg']

def get_vader_positive(text):
    return sia.polarity_scores(text)['pos']

def get_vader_neutral(text):
    return sia.polarity_scores(text)['neu']
```

```python
#Apply sentiment feature to data
drugreview_df['polarity'] = drugreview_df['Reviews'].apply(get_textblob_polarity)
drugreview_df['subjectivity'] = drugreview_df['Reviews'].apply(get_textblob_subjectivity)

drugreview_df['vader_score'] = drugreview_df['Reviews'].apply(get_vader_sentiment)
drugreview_df['vader_neg'] = drugreview_df['Reviews'].apply(get_vader_negative)
drugreview_df['vader_pos'] = drugreview_df['Reviews'].apply(get_vader_positive)
drugreview_df['vader_neu'] = drugreview_df['Reviews'].apply(get_vader_neutral)
drugreview_df['cleaned_review'] = drugreview_df['Reviews'].apply(clean_text)
```

```python
#define function that categorise reviews by sentiment
def sentiment_category(score):
    if score >= 0.5:
        return 'Highly Positive'
    elif score >= 0.1:
        return 'Positive'
    elif score <= -0.5:
        return 'Highly Negative'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

drugreview_df['sentiment_label'] = drugreview_df['vader_score'].apply(sentiment_category)

drugreview_df['sentiment_label'].value_counts()
```

### Preparing the side effect column 'Sides'

```python
#define function to clean sides column
def clean_sides(text):
    if pd.isna(text):
        return None
    
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9, ]', '', text)   # keep only letters, numbers, commas
    text = re.sub(r'\s+', ' ', text)             # normalize spaces
    text = text.replace(' ,', ',').strip()       # fix comma spacing
    return text

drugreview_df['Sides_clean'] = drugreview_df['Sides'].apply(clean_sides)
drugreview_df[['Sides', 'Sides_clean']].head(10)
```

```python
#Normalise Multiple Side Effects into Lists
drugreview_df['Sides_list'] = drugreview_df['Sides_clean'].apply(lambda x: [s.strip() for s in str(x).split(',')] if pd.notna(x) else [])
drugreview_df[['Sides_clean','Sides_list']].head(10)
```

```python
# Update modelling_df
modelling_df['Sides_clean'] = drugreview_df['Sides_clean']
modelling_df['Sides_list'] = drugreview_df['Sides_list']
print(modelling_df.head)
```

```python
#Convert Side Effects into Multi-Label Binary Columns
from sklearn.preprocessing import MultiLabelBinarizer

#define function to clean the side list column
def to_side_list(x):
    if isinstance(x, list):
        return x
    if pd.isna(x):
        return []
    # if it's a string like "nausea, headache, dizziness"
    return [s.strip() for s in str(x).split(",")]

modelling_df["Sides_list_clean"] = modelling_df["Sides_list"].apply(to_side_list)

mlb = MultiLabelBinarizer()
side_effect_encoded = mlb.fit_transform(modelling_df["Sides_list_clean"])

side_effect_df = pd.DataFrame(side_effect_encoded, columns=mlb.classes_)

modelling_df = pd.concat([modelling_df.drop(columns=["Sides_list_clean"]), side_effect_df], axis=1)
print(modelling_df.head())
```

# Classification Modelling

### Satisfaction Prediction

```python
modelling_df['Effectiveness'] = drugreview_df['Effectiveness']
modelling_df['Ease of Use'] = drugreview_df['EaseofUse']
modelling_df['Sex'] = drugreview_df['Sex']
modelling_df['Age'] = drugreview_df['Age']
```

```python
#Ensure validity of columns for modelling
modelling_df['Sex'] = modelling_df['Sex'].map({'Male': 1, 'Female': 0})
for col in ['Effectiveness', 'Ease of Use']:
    modelling_df[col] = pd.to_numeric(modelling_df[col], errors='coerce')

#Select Predictors
sentiment_vars = ['Effectiveness', 'Ease of Use', 'Sex', 'vader_score', 'vader_pos', 'vader_neg', 'vader_neu', 'polarity', 'subjectivity']

side_effect_vars = list(mlb.classes_)

X = modelling_df[sentiment_vars + side_effect_vars].fillna(0)
y = modelling_df['satisfaction_class']
```

```python
#Train and Test Split(75/25)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

#Train model
from sklearn.ensemble import RandomForestClassifier

model_class = RandomForestClassifier(
    n_estimators=300,
    random_state=42,
    class_weight='balanced'
)

model_class.fit(X_train, y_train)
```

```python
from pathlib import Path
import joblib

Path("outputs/models").mkdir(parents=True, exist_ok=True)
Path("outputs/data").mkdir(parents=True, exist_ok=True)

joblib.dump(model_class, "outputs/models/satisfaction_model.pkl")

pd.DataFrame(X_test).to_csv("outputs/data/X_test.csv", index=False)
pd.Series(y_test, name="satisfaction_class").to_csv(
    "outputs/data/y_test.csv", index=False
)
```

```python
# Evaluate the model
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

y_pred = model_class.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
#Confusion matrix plot
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, cmap="Blues", fmt='d')
plt.title("Confusion Matrix  Satisfaction Classification")
plt.xlabel("Predicted")
plt.ylabel("Actual")
save_and_close("outputs/satisfaction_confusion_matrix.png")
```

```python
#Identify Top Predictors
importances = model_class.feature_importances_
importance_df = pd.DataFrame({
    'feature': X.columns,
    'importance': importances
}).sort_values(by='importance', ascending=False)

plt.figure(figsize=(8,6))
sns.barplot(y=importance_df.head(20)['feature'], x=importance_df.head(20)['importance'])
plt.title("Top Predictors of Satisfaction")
save_and_close("outputs/satisfaction_predictors.png")
```

## Predicting Effectiveness

```python
#create effectiveness groups
modelling_df = modelling_df.dropna(subset=['Effectiveness'])

def effectiveness_group(score):
    if score >= 4:
        return "High"
    elif score == 3:
        return "Medium"
    else:
        return "Low"

modelling_df['effectiveness_class'] = modelling_df['Effectiveness'].apply(effectiveness_group)
modelling_df[['Effectiveness', 'effectiveness_class']].head(10)
```

```python
modelling_df['satisfaction_class'] = modelling_df['satisfaction_class'].map({'High': 2, 'Medium': 1, 'Low': 0})
#assign values to x, y
predictors = ['Satisfaction','Ease of Use', 'Sex', 'vader_score', 'vader_pos', 'vader_neg', 'vader_neu', 'polarity', 'subjectivity']

X = modelling_df[predictors]
#Convert effectiveness to binary
y = modelling_df['effectiveness_class']

X = X.fillna(0)
y = y.fillna(y.mode()[0])
```
